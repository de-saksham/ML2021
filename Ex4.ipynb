```
import matplotlib.pyplot as plt
import numpy as np
import gzip, pickle, sys 

f=gzip.open('MNIST.pkl.gz')
data = pickle.load(f) ;
traind = data['data_train']
traind = traind.reshape(-1,28,28)
trainl = data['labels_train']

#1
#a
print('Shape of train data is', traind.shape)
print('Shape of train lables is', trainl.shape)
print('Nr samples:', traind.shape[0])
print('Nr samples:', trainl.shape[0])

#b
#print('Sample 1000', traind[999], traind[999].shape)
sample1000 = traind[999, :, :]
plt.imshow(sample1000.reshape(28,28))
plt.show()

#c
plt.hist(trainl.argmax(axis=1))
plt.show()

#d
class5 = trainl.argmax(axis=1)
print("sample of class 5", class5)


#e
x = np.arange(0,100)
y = trainl.argmax(axis=1)
plt.bar(x, y[0:100])
plt.show()

#f
sample10 = traind[9].reshape(28,28)
flipx_ = sample10[:,::-1]
flipy_ = sample10[::-1,:]
fig = plt.figure()

fig.add_subplot(1,3,1)
plt.imshow(sample10)

fig.add_subplot(1,3,2)
plt.imshow(flipx_)

fig.add_subplot(1,3,3)
plt.imshow(flipy_)
plt.show()

#g
labels = trainl.argmax(axis=1)
class4 = (labels == 4)
data = traind[class4]
fig = plt.figure()
for i in range(0, 4):
  fig.add_subplot(1,4,i+1)
  plt.imshow(data[-i].reshape(28, 28))
plt.show()

#h
random = np.random.randint(0,100,[100])
mask = traind[random]
fig = plt.figure()
for i in range(0,3):
  fig.add_subplot(1,3,i+1)
  plt.imshow(mask[i].reshape(28,28))
plt.show()



```
